Write a Python function to calculate the Probability Density Function (PDF) of a normal distribution with mean = 50 and standard deviation = 10 at x = 65. Also plot the complete bell curve and highlight the point x = 65 on the graph.

```python
import scipy.stats as stats
def calculate_pdf(x,mean,std_dev):
  ab=stats.norm(mean,std_dev)
  return ab.pdf(x)

mean = 50
std_dev = 10
x = 65
pdf_at_point = calculate_pdf(x, mean, std_dev)
print(f"PDF at point {x} = {pdf_at_point}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
data=np.random.normal(50,10,1000)
plt.hist(data, bins=65)
```

Write a Python function to calculate the Cumulative Distribution Function (CDF) of a normal distribution with mean = 100 and standard deviation = 15 at x = 120. Also print the probability that a value is less than or equal to 120.

```python
import scipy.stats as stats
def calculate_cdf(x,mean,std_dev):
  ab=stats.norm(mean,std_dev)
  return ab.cdf(x)

mean = 100
std_dev = 15
x = 120
cdf_at_point = calculate_cdf(x, mean, std_dev)
print(f"CDF at point {x} = {cdf_at_point}")
print(f"The probality that a value is less than or equal to {x} is {cdf_at_point:.4f}")
```

```python
import scipy.stats as stat
import pylab
import numpy as np

import seaborn as sns

df=sns.load_dataset('iris')
df.head()

```

```python
import matplotlib.pyplot as plt
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(df[feature],kde=True)
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(df,'sepal_width')
```

```python
plot_data(df,'petal_width')
```

```python
plot_data(df,'sepal_length')
```

```python
plot_data(df,'petal_length')
```

Q5. Generate a random sample of size 1000 from a binomial distribution with probability of
success 0.4
and plot a histogram of the results using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.binomial(n=10, p=0.4, size=1000)
plt.hist(data, bins=range(12), edgecolor='black')
plt.title("Binomial Distribution Histogram")
plt.show()
```

Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and
calculate the
sample mean and variance.

```python
data = np.random.poisson(lam=5, size=1000)
print("Sample Mean:", np.mean(data))
print("Sample Variance:", np.var(data))
```
