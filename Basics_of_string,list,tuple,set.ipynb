48. Remove the nth word from a string.  

```python
s = input("Enter a sentence: ")
n = int(input("Enter word position to remove: "))
words = s.split()
if n >= 1 and n <= len(words):
    words.pop(n - 1)
    result = " ".join(words)
    print("After removing:", result)
else:
    print("Invalid position")
```
