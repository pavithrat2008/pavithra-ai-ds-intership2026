Q2. Write a code to print the data type of each and every element of both the variables list_ and
arra_list.

```python
import numpy as np
list_ = ['1', '2', '3', '4', '5']
array_list = np.array(list_)
print("Data types in list_:")
for i in list_:
    print(i, ":", type(i))
print()
print("Data types in array_list:")
for i in array_list:
    print(i, ":", type(i))
```
